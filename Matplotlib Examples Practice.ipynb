```python
x = np.arange(0, 2*np.pi, 0.2)
y = np.sin(x)
plt.title('Sine Wave Form')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.4)
y = np.cos(x)
plt.title('Cosine Wave Form')
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 4 * np.pi, 0.4) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 2
# and set the first such subplot as active. 
plt.subplot(2,2,1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2,2,2)
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
x = [2,8,10] 
y = [11,13,21]  

x2 = [3,9,11] 
y2 = [6,14,18] 

plt.bar(x, y) 
plt.bar(x2, y2, color = 'r') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```
